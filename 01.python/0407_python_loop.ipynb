1~10까지의 숫자 중 모든 홀수의 합을 출력하는 프로그램을 for 문을 사용하여 작성하라.

```
합: 25
```

````python
result = 0
for i in range(1, 11):
  if i % 2==1:
    result += i
  if i == 10:
    print('합 : ', result)
````

````python
# result를 0으로 정의
# for문에서 1부터 10까지 i로 할당
# if조건문에서 i를 2로 나눴을 때의 나머지가 1 = 홀수이면 result에 1을 더해준다
# 그리고 뒤의 if문에서 i가 10일때 합을 print한다
````
